```python
import pandas as pd
from functools import reduce
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('DATA_PULL_BB_72920_HARDCODE.xlsx', sheet_name='Analysis')
```

```python
list_of_df = []
for i in range(1,71,3):
    df_temp = df.iloc[:,i:i+2]
    df_temp.columns = ['Dates',df_temp.iloc[3,1]] 
    df_temp = df_temp.iloc[8:].dropna()
    df_temp['Dates'] = pd.to_datetime(df_temp['Dates'])
    list_of_df.append(df_temp)
```

```python
final_df = reduce(lambda x, y: pd.merge(x, y, how = 'left', on = 'Dates'), list_of_df)
```

```python
final_df = final_df.fillna(method='ffill')
final_df = final_df.reindex(sorted(final_df.columns), axis=1)
final_df = final_df.set_index('Dates')
```
