```python
import pandas as pd
from functools import reduce
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('Economic Date_Hardcode.xlsx', sheet_name='Analysis').iloc[:,:-16]
df2 = pd.read_excel('Economic Date_YIELDS_HARDCODED.xlsx', sheet_name='Analysis')
```

```python
list_of_df = []
for i in range(1,50,3):
    df_temp = df.iloc[:,i:i+2].dropna()
    df_temp.columns = ['Dates',df_temp.iloc[1,1] + ' [' + df_temp.iloc[2,1] + ']'] 
    df_temp = df_temp.iloc[3:]
    list_of_df.append(df_temp)
for i in range(1,16,3):
    df_temp = df2.iloc[:,i:i+2].dropna()
    df_temp.columns = ['Dates',df_temp.iloc[1,1] + ' [' + df_temp.iloc[2,1] + ']'] 
    df_temp = df_temp.iloc[3:]
    list_of_df.append(df_temp)
```

```python
final_df = reduce(lambda x, y: pd.merge(x, y, how = 'left', on = 'Dates'), list_of_df)
```

```python
final_df = final_df.fillna(method='ffill')
final_df = final_df.reindex(sorted(final_df.columns), axis=1)
final_df = final_df.set_index('Dates')
```
